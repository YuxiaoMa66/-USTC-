```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
```

```python
plt.title('Acc')  # 折线图标题
plt.rcParams['font.sans-serif'] = ['SimHei']  # 显示汉字
plt.xlabel('epoch')  # x轴标题
plt.ylabel('Accurance')  # y轴标题
plt.plot(df['epoch'], df['acc1'], marker='o', markersize=1)  # 绘制折线图，添加数据点，设置点的大小
plt.plot(df['epoch'], df['acc2'], marker='o', markersize=1)
plt.plot(df['epoch'], df['acc3'], marker='o', markersize=1)
plt.plot(df['epoch'], df['acc4'], marker='o', markersize=1)
plt.plot(df['epoch'], df['acc5'], marker='o', markersize=1)
plt.plot(df['epoch'], df['acc6'], marker='o', markersize=1)

# for a, b in zip(df['epoch'], df['acc1']):
#     plt.text(a, b, b, ha='center', va='bottom', fontsize=10)  # 设置数据标签位置及大小
# for a, b in zip(df['epoch'], df['acc2']):
#     plt.text(a, b, b, ha='center', va='bottom', fontsize=10)
# for a, b in zip(df['epoch'], df['acc3']):
#     plt.text(a, b, b, ha='center', va='bottom', fontsize=10)
# for a, b in zip(df['epoch'], df['acc4']):
#     plt.text(a, b, b, ha='center', va='bottom', fontsize=10)
# for a, b in zip(df['epoch'], df['acc5']):
#     plt.text(a, b, b, ha='center', va='bottom', fontsize=10)
# for a, b in zip(df['epoch'], df['acc6']):
#     plt.text(a, b, b, ha='center', va='bottom', fontsize=10)

plt.legend(['256隐藏，单RNN', '256隐藏，2RNN', '512隐藏，单RNN', '1024隐藏，单RNN', '256隐藏，LSTM','256隐藏，GRU'])  # 设置折线名称
plt.savefig('acc.pdf')
plt.show()  # 显示折线图
```
